Here we can see that their are 14309 rows of data and 46 columns given by the dim(data) and colnames(data) functions. I used these functions in order to see an overview of how the data is divided so that I can better understand how to perform analysis on the data.
### For each year, from 2019-2023, we are given a set of metrics.
- Tot_Spndng_YYYY: Total spending ($) for this drug in Medicare Part D.

- Tot_Dsg_Unts_YYYY: Total dosage units dispensed.

- Tot_Clms_YYYY: Total number of prescription claims.

- Tot_Benes_YYYY: Total number of Medicare beneficiaries who used the drug.

- Avg_Spnd_Per_Dsg_Unt_Wghtd_YYYY: Average spending per dosage unit (weighted).

- Avg_Spnd_Per_Clm_YYYY: Average spending per claim.

- Avg_Spnd_Per_Bene_YYYY: Average spending per beneficiary.

- Outlier_Flag_YYYY: Whether CMS flagged this drug as an outlier in that year (e.g., unusual growth or spending).

### We are also given two trend columns:
- Chg_Avg_Spnd_Per_Dsg_Unt_22_23: The change in average spending per dosage unit between 2022 and 2023.

- CAGR_Avg_Spnd_Per_Dsg_Unt_19_23: Compound Annual Growth Rate of spending per dosage unit from 2019 to 2023.

```r
overall <- data %>% filter(Mftr_Name == "Overall")

top10_2019 <- overall %>%
  arrange(desc(Tot_Clms_2019)) %>%
  select(Brnd_Name, Gnrc_Name, Tot_Clms_2019) %>%
  head(10)

top10_2020 <- overall %>%
  arrange(desc(Tot_Clms_2020)) %>%
  select(Brnd_Name, Gnrc_Name, Tot_Clms_2020) %>%
  head(10)

top10_2021 <- overall %>%
  arrange(desc(Tot_Clms_2021)) %>%
  select(Brnd_Name, Gnrc_Name, Tot_Clms_2021) %>%
  head(10)

top10_2022 <- overall %>%
  arrange(desc(Tot_Clms_2022)) %>%
  select(Brnd_Name, Gnrc_Name, Tot_Clms_2022) %>%
  head(10)

top10_2023 <- overall %>%
  arrange(desc(Tot_Clms_2023)) %>%
  select(Brnd_Name, Gnrc_Name, Tot_Clms_2023) %>%
  head(10)

top10_2019
top10_2020
top10_2021
top10_2022
top10_2023
```

### Code Explanation
- In the code above, I am filtering the dataset to keep only the "Overall" rows. These rows represent the total number of claims across all the manufactures of each drug.
- I then sort, for each year 2019-2023, the top 10 drugs by total claims in descending order.
- I then display these top 10 drugs in separate tables, this allows me to compare how the list of most common drugs has potentially changed over time.

### Interpreting the Results
- We can see in our results that the top 10 drugs for each year is actually very consistent, which supports my hypothesis.
- From 2019-2022, the top 10 drugs are identical: Atorvastatin Calcium, Amlodipine Besylate, Lisinopril, Levothyroxine Sodium, Gabapentin, Omeprazole, Losartan Potassium, Furosemide, Metformin HCl, and Metoprolol Succinate are all the most commonly prescribed.
- In 2023, there is a small change, where Rosuvastatin Calcium enters the top 10, displacing Furosemide.
- After some reflection and research this does make sense, as many of these drugs are cholesterol, blood pressure, and heart medicines, which relate to chronic diseases often found in the Medicare population.

### How this answers the question
- This answers the first part of my question, "What drugs are most commonly prescribed in Medicare".

```r
mftr_top_2019 <- data %>%
  filter(Mftr_Name != "Overall",
         Brnd_Name %in% top10_2019$Brnd_Name) %>%
  group_by(Mftr_Name) %>%
  summarise(Total_Spndng_2019 = sum(Tot_Spndng_2019, na.rm = TRUE), .groups = "drop") %>%
  arrange(desc(Total_Spndng_2019)) %>%
  head(10)

mftr_top_2020 <- data %>%
  filter(Mftr_Name != "Overall",
         Brnd_Name %in% top10_2020$Brnd_Name) %>%
  group_by(Mftr_Name) %>%
  summarise(Total_Spndng_2020 = sum(Tot_Spndng_2020, na.rm = TRUE), .groups = "drop") %>%
  arrange(desc(Total_Spndng_2020)) %>%
  head(10)

mftr_top_2021 <- data %>%
  filter(Mftr_Name != "Overall",
         Brnd_Name %in% top10_2021$Brnd_Name) %>%
  group_by(Mftr_Name) %>%
  summarise(Total_Spndng_2021 = sum(Tot_Spndng_2021, na.rm = TRUE), .groups = "drop") %>%
  arrange(desc(Total_Spndng_2021)) %>%
  head(10)

mftr_top_2022 <- data %>%
  filter(Mftr_Name != "Overall",
         Brnd_Name %in% top10_2022$Brnd_Name) %>%
  group_by(Mftr_Name) %>%
  summarise(Total_Spndng_2022 = sum(Tot_Spndng_2022, na.rm = TRUE), .groups = "drop") %>%
  arrange(desc(Total_Spndng_2022)) %>%
  head(10)

mftr_top_2023 <- data %>%
  filter(Mftr_Name != "Overall",
         Brnd_Name %in% top10_2023$Brnd_Name) %>%
  group_by(Mftr_Name) %>%
  summarise(Total_Spndng_2023 = sum(Tot_Spndng_2023, na.rm = TRUE), .groups = "drop") %>%
  arrange(desc(Total_Spndng_2023)) %>%
  head(10)

mftr_top_2019
mftr_top_2020
mftr_top_2021
mftr_top_2022
mftr_top_2023
```

### Code Explanation
- I start by taking the top 10 most prescribed drugs from my previous code above, and filter the data such that I exclude the "Overall" rows and instead just have the manufacturer rows.
- I then sum the total spending by the manufacturing name for the top 10 drugs, and ignore empty data. Unfortunately, this data does not contain revenue data, so the total spending is the closest data type I can use.
- I then sort the findings in descending order and display it in a table.
- I do this so that it will be clear which manufactures spend the most, which may give some insight to which companies are profiting the most.
### Interpreting the Results
- Across 2019-2023 we can see that the most spending on these top 10 drugs is done by the same companies (e.g Mylan, Lupin, Dr.Reddy'S, Aurobindo, Amneal, Apotex, Zydus, Ascend, and Northstar appear every year). However these manufacturers' rankings change over the years. For example, Mylan loses its spot as the biggest spender after 2020, and new manufactures (Glenmark, Camber) appear in 2023.
- Looking at the spending trends, we can see Mylan (originally dominating in the spending sector in 2019), dramatically decreased their spending from 560 million dollars to 292 million dollars. We can also see Lupin having a huge increase in spending, from 342 million dollars to 442 million dollars. The rest of the manufacturers, while having some fluctuations, remained relatively stable.
- Through this data, we observe a major market leadership shift from Mylan to Lupin, as well as more even spending amonst the top-10. We also see new manufactures emerging, which suggests a more competitive market in 2023.

```r
overall <- data %>% filter(Mftr_Name == "Overall")
change <- overall %>%
  filter(Brnd_Name %in% top10_2023$Brnd_Name) %>%
  select(
    Brnd_Name, Gnrc_Name,
    Chg_Avg_Spnd_Per_Dsg_Unt_22_23,
    CAGR_Avg_Spnd_Per_Dsg_Unt_19_23
  )
  change
```

### Code Explanation
- I take the top-10 drugs from 2023, and filter them to show both the one year change in average spending per dosage unit from 2022 to 2023, as well as the compound annual growth rate of per-unit spending from 2019 to 2023.
- This will allow me to easily analyze the changes in spending on the dosage per unit.

### Interpreting the Results

#### Change from 2022 to 2023
- We can see an increase of spending per dosage unit from Metformin HCI (+26.7%), Lisinopril (+21.4%), Losartan Potassium (+20.4%), Atorvastatin (+10.6%), and Omeprazole (+3.4%).
- We can see a decrease in spending per dosage unit from Rosuvastatin (−16.9%), Levothyroxine (−10.2%), Gabapentin (−6.8%), and Metoprolol (−0.9%).

#### Compound Annual Growth Rate from 2019-2023
- We can see an increase for Losartan (+5.4%/yr), Metformin (+4.1%/yr), Lisinopril (+2.4%/yr), and Amlodipine (+1.5%/yr).
- We can see a decrease for Rosuvastatin (−17.7%/yr), Levothyroxine (−8.7%/yr), Metoprolol (−10.6%/yr), Atorvastatin (−5.4%/yr), Gabapentin (−1.7%/yr), and Omeprazole (−1.5%/yr)

#### Conclusion
- The set of most-prescribed drugs are stable, but their unit costs are not.
- Generic drugs used for blood pressure and diabetes show consistent increases in spending per unit, while generic drugs used for cholesterol show flat to decreased price per unit.
- This means that while the volume of these drugs being sold maintains relatively stable, the price per unit fluctuates.

### How this answers my question
- This directly answers my question, "How has the price of these drugs changed throughout these years". The results have been shown to support my hypothesis that the prices will fluctuate dramatically throughout the pandemic.
